```python

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 250
import matplotlib.pyplot as plt
import AnalysisEngine
import numpy as np

from scipy import fft

from analysis_engine.simulator import data_generator, add_field, modify_resolution, projections
from analysis_engine.utils import data_reader, plotting
from analysis_engine.statistics import statistics_base
from analysis_engine.statistics.spectra import spectra_base, per_spectra, arevalo_spectra, flatsky_spectra
from analysis_engine.statistics.statfunc import statfunc_base, corr, strfn, statfunc_mpi_naive

c = plotting.CB_color_cycle
```

```python
_pl = -11./3.
_n = 256
L = 2.*np.pi
L2 = [L, L]
## GAUSSIAN FLUCTUATION FIELD
x2 = data_generator.generate_field((_n, _n), [2., -11./3.], [2.])
```

```python
## PERCENTAGE OF DATA TO NOT MASK OUT
P = 0.5
r_cut = np.sqrt(P * _n**2 / np.pi)
```

```python
## RANDOM CIRCLES MASK
m0 = data_generator.generate_mask(x2.shape, p=P)

## MASK OUTSIDE REGION
m1 = np.ones_like(x2)
grid = np.indices(x2.shape)
grid = [grid[i,...] - n//2 for i, n in enumerate(x2.shape)]
idx = np.linalg.norm(grid, axis=0) > r_cut
m1[idx] = 0

## USE GAUSSIAN FIELD TO GENERATE A MASK
m2_p = 1.
val = 0.
while m2_p > P:
    m2 = data_generator.generate_field((_n, _n), [2., -11./3.], [2.], seed=423)
    m2[np.abs(m2) < val] = 0
    m2[np.abs(m2) >= val] = 1
    m2_p = np.sum(m2)/np.prod(m0.shape)
    val = val + 0.2

## PLOT THE FLUCTUATIONS WITH THE MASK
fig, ax = plotting.get_fig((1, 3))

## setting masked regions to `nan` only for plotting purposes
pl0 = x2*m0
pl0[m0 == 0] = np.nan
pl1 = x2*m1
pl1[m1 == 0] = np.nan
pl2 = x2*m2
pl2[m2 == 0] = np.nan

cmap = plotting.make_bad_cmap('bwr')
cmin, cmax = plotting.get_symmetric_minmax(x2)
ax_kwargs = {'interpolation': 'none', 'origin': 'lower', 'vmin': cmin, 'vmax': cmax, 'cmap': cmap}
plotting.imshow_cbar(pl0, fig, ax[0], ax_kwargs)
plotting.imshow_cbar(pl1, fig, ax[1], ax_kwargs)
plotting.imshow_cbar(pl2, fig, ax[2], ax_kwargs)

for a in ax:
    a.set_yticks([])
    a.set_xticks([])

print(np.sum(m0)/np.prod(m0.shape), np.sum(m1)/np.prod(m1.shape), np.sum(m2)/np.prod(m2.shape))
```

```python
## FLATSKY SPECTRUM
kf0, fekf0 = flatsky_spectra.modal_spectrum(x2, mask=m0, lenn=L2)
kf1, fekf1 = flatsky_spectra.modal_spectrum(x2, mask=m1, lenn=L2)
kf2, fekf2 = flatsky_spectra.modal_spectrum(x2, mask=m2, lenn=L2)

## THE TRUE SPECTRUM (NO MASK FLUCTUATIONS)
k, fek, _ = spectra_base.calculate_integrated_spectrum((x2,), spec_type='modal', lenn=L2, int_kwargs={'bin_center': True})

## GET AREVALO METHOD TO PRODUCE SAME (EQUIVALENT) WAVENUMBERS
dsig = (np.sqrt(2)/np.pi) * (L / _n)
lags = np.sqrt(2) / (k * dsig)

## AREVALO METHOD
## NOTE: supply arevalo method with the mask AND multiply the fluctuations with the mask
ka0, feka0 = arevalo_spectra.modal_spectrum(x2*m0, exp1=m0, lags=lags, lenn=L)
ka1, feka1 = arevalo_spectra.modal_spectrum(x2*m1, exp1=m1, lags=lags, lenn=L)
ka2, feka2 = arevalo_spectra.modal_spectrum(x2*m2, exp1=m2, lags=lags, lenn=L)

fig, ax = plotting.get_fig((1, 1))
ax.loglog(k, fek, color='black')

ax.loglog(kf0, fekf0, color=c[0], linestyle='--')
ax.loglog(kf1, fekf1, color=c[1], linestyle='--')
ax.loglog(kf2, fekf2, color=c[2], linestyle='--')

ax.loglog(ka0, feka0, color=c[0], linestyle=':')
ax.loglog(ka1, feka1, color=c[1], linestyle=':')
ax.loglog(ka2, feka2, color=c[2], linestyle=':')

ax.plot([], [], color='black', label='True (no mask)')
ax.plot([], [], color=c[0], label='Random mask')
ax.plot([], [], color=c[1], label='Outside region masked')
ax.plot([], [], color=c[2], label='FBM masked')

ax.legend()
```
